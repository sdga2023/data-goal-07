```python
#!pip install wbgapi
import wbgapi as wb
import plotly.express as px
```

```python
df = wb.data.DataFrame(['EG.ELC.ACCS.ZS', 'SP.POP.TOTL'], time=2020, labels=True).reset_index()
fig = px.choropleth(df, locations="economy",
                    color="EG.ELC.ACCS.ZS",
                    hover_name="Country", # column to add to hover information
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```
